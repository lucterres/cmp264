```python
import numpy as np
```

```python
ix = 91
jy = 71
xMin = 1000000 
xMax = 1090000
yMin = 7230000
yMax = 7300000
zMin = -2607.8 
zMax = -1796.01
x = np.linspace(xMin, xMax, ix)
y = np.linspace(yMin, yMax, jy)
X,Y = np.meshgrid(x, y)

Z = np.zeros((jy, ix))
Z[:,:] = zMin
```

```python
import matplotlib.pyplot as plt
# Crie uma figura
fig = plt.figure(figsize=(10, 10))

# Adicione um grfico 3D  figura
ax = fig.add_subplot(111, projection='3d')

# Plote a superfcie
ax.plot_surface(X, Y, Z, cmap='viridis')

# Mostre o grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crie os dados para o eixo X, Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Crie os dados para o eixo Z
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Crie uma figura
fig = plt.figure(figsize=(20, 20))

# Adicione um grfico 3D  figura
ax = fig.add_subplot(111, projection='3d')

# Plote a superfcie
ax.plot_surface(X, -Z, Y, cmap='viridis')

# Mostre o grfico
plt.show()
```

```python
# Crie um scatter plot com a escala de cores definida por z
plt.scatter(X, Y, c=Z, cmap='viridis')

# Adicione uma barra de cores
plt.colorbar(label='Varivel para escala de cores')
```
